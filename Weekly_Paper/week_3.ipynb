## 1. 1종 오류와 2종 오류에 대해 설명해주세요
- **1종 오류** : 귀무가설(H0)이 실제로는 참이엇 채택해야 함에도 불구하고 표본의 오차때문에 이를 채택하지 않는 오류를 말합니다. 보통 Alpha 로 표기하고 유의수준이라 부릅니다.

- **2종 오류** : 귀무가설(H0)이 거짓이라서 채택하지 말아야 하는데 표본의 오차 때문에 이를 채택하는 오류를 말합니다. 보통 Beta로 표기합니다.

	예를 들어 한 공장에서 기계를 전부 바꿔야 한다는 주장이 나왔다고 합시다.
	이럴 경우 가설의 설정을 다음과 같이 할 수 있습니다.
        
	`H0(귀무가설) : 공장의 기계를 안 바꾸는 게 낫다.`  
	`H1(대립가설) : 공장의 기계를 바꾸는게 낫다.`

	제1종 오류의 경우  :  
	H0의 주장이 맞음에도 H1를 따르게 되어 기계를 바꿔야 하는 금액적 손실이 이뤄질 수 있습니다.

	제2종 오류의 경우  :  
	H1이 주장이 맞지만 H0을 채택하여 비록 새롭게 주장한 내용이 맞더라도 기계를 바꿔야 하는 경제적 부담은 없게 됩니다.


## 2. p값(p-value)는 무엇인가요?
- **p값(p-value)** : 관찰된 데이터가 귀무가설과 일치하는 정도를 나타내는 값  
		- P값이 높은 경우 : 귀무가설이 참일 가능성이 높은 데이터입니다.  
		- P값이 낮은 경우 : 귀무가설이 참일 가능성이 낮은 데이터입니다.

- p-value는 어떤 사건(관찰된 값)이 우연히 발생할 확률이라고 말할 수도 있다. 즉, p값이 `0.05`보다 작다는 말은 어떤 사건이 우연히 발생할 확률이 5%보다 작다는 의미이다. 따라서 이 사건이 우연히 일어났을 가능성이 거의 없다고 추정하는  것이다. 즉 이 사건은 우연히 발생한 것이 아니라 '무언가 인과관계가 있기 때문에 발생한 것일거다. 이정도면 우연은 아닐거다'라고 해석할 수 있다. 반대로 p값이 `0.05`보다 크면 어떤 사건이 우연히 발생할 확률이 5%보다 크기 때문에, 해당 사건은 통계적으로 인과관계가 없다고 해석할 수 있다. 이렇게 설명하면 '그 사건이 우연히 일어났다'라고 가정하는 것이 귀무가설이고 '그 사건은 우연히 일어난것이 아니고 뭔가 이유가 있다'라고 가정하는 것이 대립가설이다. 따라서 따라서 p-value가 5%보다 작다면 우연히 일어날 가능성이 거의 없기 때문에 귀무가설을 기각하고 대립가설을 채택하는 것이다.
