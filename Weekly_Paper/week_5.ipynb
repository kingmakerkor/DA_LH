## 사분위수에 대해 설명해 주세요.  
- 사분위수는 데이터의 분포를 이해하고, 중심 경향과 변동성을 평가하는데 유용한 통계적 범위를 일컫음  
  
#### 사분위수의 종류
- 제1사분위수(Q1) : 데이터의 하위 25%가 이 값보다 작거나 같습니다. 데이터의 첫 번째 1/4 지점에 해당하는 값.
- 제2사분위수(Q2) : 데이터의 하위 50%가 이 값보다 작거나 같은데, 이는 중앙값과 동일.
- 제3사분위수(Q3) : 데이터의 하위 75%가 이 값보다 작거나 같습니다. 데이터의 세 번째 1/4 지점에 해당하는 값.
  
#### 사분위간 범위(IQR = Q3 -Q1)
- 사분위간 범위는 제3사분위수(Q3)에서 제1사분위수(Q1)를 뺀 값으로, 데이터의 중간 50%가 분포하는 범위.
- 이상치를 탐지하거나 데이터의 변동성을 평가하는데 사용
  
#### 특징 및 활용
- 사분위수는 데이터 집합의 중앙 경향을 파악하는데 유용
- IQR은 이상치를 탐지할 때 유용
- Q1 - (1.5 * IQR) 보다 작거나 Q3 + (1.5 * IQR)보다 큰 값을 이상치로 간주<br><br><br>
  
## 기술통계와 추론통계는 무엇이고, 어떤 차이가 있나요?  
#### 1. 기술통계  
- 간단한 수치나 그래프를 통해 데이터를 직관적으로 이해할 수 있도록 표현
- 데이터의 전체적인 경향성과 분포를 파악하는 데 유용
- 주요 개념
    - 중심 경향성 : 데이터의 중심값을 나타내는 지표(평균, 중앙값, 최빈값)
    - 산포도 : 데이터가 얼마나 퍼져 있는지를 나타내는 지표(분산, 표준편차, 범위)
    - 분포 : 데이터가 어떻게 분포되어 있는지를 나타내는 지표(왜도, 첨도)
    - 시각화 도구 : 데이터를 시각적으로 표현(히스토그램, 박스플롯, 도수분포표 등)

#### 2. 추론통계
- 표본 데이터를 바탕으로 모집단에 대한 결론을 도출하는 통계 기법
- 전체 모집단에서 일부 표본만을 가지고 그 표본이 모집단을 얼마나 잘 대표하는지 평가
- 확률 이론에 기반하여 불확실성을 고려하면서 결론을 도출
- 주요 개념
    - 표본과 모집단
        - 표본 : 모집단에서 선택된 일부 집합
        - 모집단 : 연구 대상이 되는 전체 집합
    - 추정
        - 점 추정 : 모집단의 특정 모수를 하나의 값으로 추정하는 방법
        - 구간 추정 : 모수를 하나의 값이 아닌 범위로 추정하는 방법
    - 가설검정 : 특정 가설을 설정하고 표본 데이터를 통해 그 가설이 맞는지 검증
        - 귀무가설 : 차이가 없거나 효과가 없다는 가설
        - 대립가설 : 차이가 있거나 효과가 있다는 가설
        - p-값(p-value)을 계산하여 귀무가설을 기각할지 여부를 결정
        - p-값이 작으면 귀무가설을 기각하고 대립가설을 채택합니다(보통 p < 0.05일 때)
    - 상관관계와 회귀분석
        - 상관관계 : 두 변수 간의 관계를 파악하는 방법
        - 회귀분석 : 독립변수가 종속변수에 미치는 영향을 분석하여 예측 모델을 만드는 방법  

#### 3. 기술통계와 추론통계 차이<br>
|구분|기술통계|추론통계|
|------|---|---|
|목적|주어진 데이터를 요약하고 설명|표본 데이터를 통해 모집단에 대한 결론 도출|
|대상|전체 데이터(모집단 또는 표본)|표본 데이터|
|결과|데이터 자체에 대한 요착 정보 제공|모집단에 대한 추정 및 예측|
|방법|평균,중앙값,최빈값 등 중심 경향성 및 산포도 분석|가설검정, 신뢰구간 설정 및 모두 추정|
|활용 사례|특정 집단의 성적 평균 계산 및 분포 시각화|약물 실험 결과를 바탕으로 전체 인구에게 효과를 예측|
