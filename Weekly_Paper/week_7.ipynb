# 1. 장바구니 분석의 활용 사례를 설명해주세요. 해당 사례에서 얻을 수 있는 비즈니스 인사이트는 무엇인지 구체적으로 설명해 주세요

### 제품 배치 최적화
- `활용 사례`: 슈퍼마켓에서 장바구니 분석을 통해 자주 함께 구매되는 제품을 파악하여, 물리적으로 가까운 위치에 배치합니다.  
- `비즈니스 인사이트`: 고객이 자주 함께 구매하는 제품을 인접하게 배치하면, 고객의 쇼핑 편의성을 높이고 충동 구매를 유도할 수 있습니다.  
- `예시)` 빵과 우유를 가까운 위치에 배치함으로써 두 제품의 판매를 촉진할 수 있습니다.  
### 크로스 셀링(Cross-Selling)  
- `활용 사례`: 전자상거래 사이트에서 특정 제품을 구매한 고객이 자주 함께 구매하는 다른 제품을 파악하여 추천합니다.  
- `비즈니스 인사이트`: 고객이 특정 제품을 구매할 때 관련 제품을 추천하여 추가 구매를 유도할 수 있습니다.  
- `예시)` 노트북을 구매한 고객에게 마우스나 키보드를 추천하는 전략을 통해 매출을 증대시킬 수 있습니다.  
### 프로모션 전략 수립  
- `활용 사례`: 장바구니 분석을 통해 특정 제품 조합에 대한 프로모션을 기획합니다.  
- `비즈니스 인사이트`: 자주 함께 구매되는 제품에 대해 번들 할인이나 패키지 상품을 구성하여 제공함으로써, 고객에게 더 큰 가치를 제공하고 매출을 증가시킬 수 있습니다.  
- `예시)` 커피와 도넛을 함께 할인된 가격으로 제공하는 프로모션이 가능합니다.  
### 재고 관리 최적화  
- `활용 사례`: 자주 함께 구매되는 제품 간의 수요를 예측하여 재고를 효율적으로 관리합니다.  
- `비즈니스 인사이트`: 재고 부족이나 과잉을 방지할 수 있으며, 효율적인 재고 관리를 통해 비용 절감과 매출 증대를 도모할 수 있습니다.  
### 고객 충성도 강화  
- `활용 사례`: 고객이 자주 구매하는 제품 조합을 이해하여 맞춤형 로열티 프로그램이나 특별 할인을 제공합니다.  
- `비즈니스 인사이트`: 고객의 기대와 요구를 충족시켜 만족도와 충성도를 높일 수 있습니다. 이는 장기적으로 고객 유지율 향상과 브랜드 신뢰도 강화로 이어질 수 있습니다.  

# 2. Support, Confidence, Lift 지표의 정의와 각 지표의 중요성을 설명해 주세요. 이 지표들을 해석하는 방법을 구체적인 예와 함께 설명해 주세요.

### 신뢰도(confidence) 의 의미  
- 정의: 해당 아이템이 포함된 거래 중에서 다른 아이템도 함께 구매할 경우를 보여 주는 지표이며, 다른 지표와의 다르게 방향성의 의미를 가지고  있어, 방향성에 따라 신뢰도가 달라지는 경향을 보여준다  
- 중요성: 높은 지지도는 해당 아이템이 자주 발생함을 의미하며, 분석에 있어 중요한 패턴을 찾아내는 데 도움을 줌

### 지지도(support)의 의미  
- 정의: 해당 아이템이 전체 거래 데이터에서 얼마나 자주 등장하는지를 보여 주는 지표이며, 아이템의 빈도를 나타냄  
- 중요성: 실질적인 행동으로 옮길 수 있는 규칙을 제공함

### 향상도(lift)의 의미  
- 정의: 아이템이 독립적으로 판매되는 경우에 비해 아이템 A와 아이템 B 가 함께 구매될 가능성이 얼마나 높은지를 측정하는 지표이며, 아이템 간의 관계 강도를 평가하는 지표이다. 1보다 크면 양의 상관관계를, 1보다 작으면 음의 상관관계를, 1과 같다면 두 아이템 간의 상관 관계는 없다로 확인한다  
- 중요성: 두 아이템이 독립적일 때보다 더 자주 함께 발생함을 나타냄

### 예시

#### 100 개의 거래 데이터
커피: 50번 구매됨 (0.5)  
도넛: 30번 구매됨 (0.3)  
커피 + 도넛: 20번 구매됨 (0.2)  

지지도  
규칙 : {커피, 도넛} = 0.2  
지지도 0.2는 전체 거래 중 20%에서 커피와 크림이 함께 구매되었다는 것을 의미  
높은 지지도는 이 조합이 자주 발생함을 나타내며, 분석에서 중요한 패턴으로 간주  

신뢰도  
규칙 : {커피} -> {도넛} = 0.4 ( 0.2 / 0.5 = 0.4)  
신뢰도 0.4는 커피를 구매한 고객 중 40%가 도넛을 함께 구매했다는 것을 의미  
신뢰도가 높을수록 두 상품 간의 관계가 강하다고 보임  
커피를 구매한 고객에게 도넛을 추천하는 전략이 어느 정도 효과적일 수 있음을 시사함  

향상도  
규칙 : {커피} -> {도넛} = 1.33 ( 0.2 / ( 0.5 * 0.3 ) = 1.33)  
향상도 1.33은 커피와 도넛이 독립적으로 발생할 때보다 약 33% 더 자주 함께 구매된다는 것을 의미  
Lift > 1: 긍정적인 연관성을 나타냄(즉, 두 상품이 서로 관련 있음)  
Lift = 1: 두 상품 간에 아무런 연관성이 없음  
Lift < 1: 부정적인 연관성을 나타냄(즉, 두 상품이 함께 구매될 가능성이 낮음)  

제안  
- 커피를 장바구니에 담은 고객에게 "도넛을 추가로 구매하시겠습니까?"라는 팝업 메시지를 표시하여 추가 구매를 유도  
- 커피와 도넛의 연관성을 고려하여 "커피와 도넛 세트 할인" 같은 번들 프로모션을 기획  
- 매장에서 도넛 품목을 캐셔 위치에 배치하여 커피가 나오는 동안 고객이 머무는 동안 제품을 함께 구매하도록 유도  
- 두 제품의 연관성을 고려하여 재고 관리
